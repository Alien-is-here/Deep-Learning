```python
img = image.load_img("/content/cat.jfif", target_size=(200, 200))
```

```python
datagen = ImageDataGenerator(
        rotation_range=30,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        width_shift_range=0.2,
        height_shift_range=0.2,
)
```

```python
img_array = image.img_to_array(img)
```

```python
img_array.shape
```

```python
inputs = img_array.reshape((1,) + img_array.shape)
```

```python
import os

# Create a directory named 'preview' if it doesn't exist
os.makedirs("/content/preview", exist_ok=True)
```

```python
i = 0
for output in datagen.flow(inputs, batch_size=1, save_to_dir='preview', save_prefix='cat', save_format='jpeg'):
  i += 1
  if i > 20:
    break
```
